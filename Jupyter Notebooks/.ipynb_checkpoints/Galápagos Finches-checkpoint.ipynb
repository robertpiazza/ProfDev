# Galápagos Island Finches

Every year for the past 40-plus years, Peter and Rosemary Grant have gone to the Galápagos island of Daphne Major and collected data on Darwin's finches. Using statistical inference, we will analyze their data and witness evolution in action! Let's get to it!
